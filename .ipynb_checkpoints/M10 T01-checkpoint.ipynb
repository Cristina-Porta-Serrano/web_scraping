### BeatifulSoap

We use the requests library to download the page content. It sends an HTTP request to the URL of the webpage you want to access. The server responds to the request by returning the HTML content of the webpage.

We create the "soup", the Python object tree that represent the HTML document. Once we have accessed the HTML content, we are left with the task of parsing the data. Since most of the HTML data is nested, we cannot extract data simply through string processing. One needs a parser which can create a nested/tree structure of the HTML data. 

```python
URL = "http://quotes.toscrape.com/tag/inspirational/" #specify the URL of the webpage you want to scrape.
URL2 = "http://quotes.toscrape.com/tag/inspirational/page/2/"
page = requests.get(URL) #Send a HTTP request to the specified URL and save the response from server in a response object 
#called page.
page2 = requests.get(URL2)
```

```python
print(page.text) #Now, print to get the raw HTML content of the webpage. It is of ‘string’ type.

soup = BeautifulSoup(page.content, 'html5lib') #Parsing the HTML content: page.content is the raw HTML content, html5lib is the HTML parser we use.

print(soup.prettify()) #it gives the visual representation of the parse tree created from the raw HTML content.
```

```python
print(page2.text) #Now, print to get the raw HTML content of the webpage. It is of ‘string’ type.

soup2 = BeautifulSoup(page2.content, 'html5lib') #Parsing the HTML content: page.content is the raw HTML content, html5lib is the HTML parser we use.

print(soup2.prettify()) #it gives the visual representation of the parse tree created from the raw HTML content.
```

```python
quotes = soup.find_all(class_ = "quote")
quotes2 = soup2.find_all(class_ = "quote")
```

```python
author = []
text = []

for i in quotes:
    text.append(i.find(class_= "text").text)
    author.append(i.find(class_ = "author").text)
```

```python
author2 = []
text2 = []

for i in quotes2:
    text2.append(i.find(class_= "text").text)
    author2.append(i.find(class_ = "author").text)
```

```python
content = driver.find_elements(by= By.XPATH, value = '//div[@class="quote"]')
```

```python
sentence = []
person = []
for element in content:
    sentence.append(element.find_element(By.CLASS_NAME, 'text').text)
    person.append(element.find_element(By.CLASS_NAME, 'author').text)
```

### BeatifulSoap

```python
URL_bolsa = "https://www.bolsamadrid.es/esp/aspx/Mercados/Precios.aspx?indice=ESI100000000" #specify the URL of the webpage you want to scrape.

page_bolsa = requests.get(URL_bolsa) #Send a HTTP request to the specified URL and save the response from server in a response object 
#called page.
```

```python
print(page_bolsa.text) #Now, print to get the raw HTML content of the webpage. It is of ‘string’ type.

soup_bolsa = BeautifulSoup(page_bolsa.content, 'html5lib') #Parsing the HTML content: page.content is the raw HTML content, html5lib is the HTML parser we use.

print(soup_bolsa.prettify()) #it gives the visual representation of the parse tree created from the raw HTML content.
```

```python
table1 = soup_bolsa.find('table', id= 'ctl00_Contenido_tblAcciones')
table1
```

```python
column_names = []

for i in table1.find_all('th'): 
    column_name = i.text
    column_names.append(column_name)
    
print(column_names)
```

```python
# create dataframe with column_names 
stock_prices = pd.DataFrame(columns=column_names)
stock_prices
```

```python
for j in table1.find_all('tr')[1:]:   # tr -> rows (begins in second row, firts is our columns names)
    row_data = j.find_all('td')       # td -> items in each row 
    row = [i.text for i in row_data]
    length = len(stock_prices)
    #  we fill in each row according to its index 
    stock_prices.loc[length] = row  
```

```python
stock_prices
```

### Selenium

```python
# we will use the driver to get access to the browser Google Chrome 
driver_bolsa = webdriver.Chrome(executable_path = 'C:/Users/Pereira/Downloads/chromedriver_win32/chromedriver.exe')

# url of the page that hosts the table 
url_bolsa = "https://www.bolsamadrid.es/esp/aspx/Mercados/Precios.aspx?indice=ESI100000000" 

# use the driver to open the url 
driver_bolsa.get(url_bolsa)
```

```python
table_bolsa = []
object = driver_bolsa.find_elements(by= By.XPATH, value = '//*[contains(@id,"ctl00_Contenido_tblAcciones")]//tr')
for element in object:
    row = [item.text for item in element.find_elements(by= By.XPATH, value= ".//*[self::td or self::th]")]
    table_bolsa.append(row)
```

```python
precios_de_la_sesion = pd.DataFrame(table_bolsa[1:], columns = table_bolsa[0])
precios_de_la_sesion
```

#### Context
This dataset contains information of the session prices of the IBEX 35 Spanish stock market. The data has been collected from The Madrid stock exchange website on 23/09/2022 at 19:00 p.m. 

#### Content
35 rows and 9 columns.
Columns' description are listed below.

- Nombre: Name of the Company.
- Últ.: Current share price (in €).
- % Dif: Percentage difference with respect to the opening price.
- Máx. : Maximum value of the price in the current session.
- Mín. : Minimum value of the price in the current session.
- Volumen : Number of transactions of the asset carried out until the close of the trading session.
- Efectivo (miles €) : Transactions carried out up to the close of the trading session, expressed in thousands of euros.
- Fecha : Date of the session.
- Hora : Time of of the day. 'Cierre' means the time is when the session is closed.
